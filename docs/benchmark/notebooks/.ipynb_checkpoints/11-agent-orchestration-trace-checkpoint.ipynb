# 11 - Agent 编排层追踪

**目标**: 测量 AI Engine 内部 Agent 编排每个节点的时间分布，量化意图识别、字段提取、文档生成各阶段的耗时。

**覆盖事件链**:
```
[AGENT-1] GreeterAgent.route() — 意图识别（LLM call #1）
  ↓ intent=procurement
[AGENT-2] RequirementAgent 字段提取（LLM call #2）
  ↓ [NODE: extract_fields]    — ConfigurableFieldExtractor + call_llm
  ↓ [NODE: check_completeness] — 验证必填字段
  ↓ (category_agent: 品类识别，基于关键词匹配，无LLM)
[AGENT-3] ClerkAgent 文档生成（LLM call #3）
  ↓ 生成结构化需求文档
[SSE] 每个节点完成后流式推送
```

**架构说明**:
- `run_requirement_agent_stream`: 手动状态机（greeter→requirement→clerk→done），非 LangGraph graph
- `RequirementAgent.build_graph()`: LangGraph StateGraph（用于非流式路径）
- `GreeterAgent`: 无状态，每次调用独立 LLM 分类
- Session 存储: 内存 dict（Phase 0），无 Redis/DB 依赖

````python
# ============================================================
# CELL 0: 配置区域 — 所有可控参数在此修改
# ============================================================

# --- 服务端点 ---
AI_ENGINE_URL  = "http://localhost:8002"   # xiaocai-ai-engine
API_GATEWAY_URL = "http://localhost:8001"  # xiaocai-api (转发层)

# --- 采集参数 ---
N_SAMPLES_LLM  = 5    # LLM 相关测试（昂贵，次数少）
N_SAMPLES_FAST = 20   # 非 LLM 测试（快速，次数多）
WARMUP_ROUNDS  = 2    # 预热轮数
TIMEOUT_S      = 120.0  # 含 LLM 调用的请求超时

# --- Session ID（每次运行用不同 ID 避免缓存影响）---
import time as _time
_ts = int(_time.time())
SESSION_GREETER_ONLY     = f"bench-greeter-{_ts}"
SESSION_PROCUREMENT_1    = f"bench-proc-1-{_ts}"
SESSION_PROCUREMENT_2    = f"bench-proc-2-{_ts}"
SESSION_EMBED            = f"bench-embed-{_ts}"

# --- 测试消息 ---
MSG_NON_PROCUREMENT = "今天天气怎么样？"           # 非采购意图（触发 GreeterAgent 拦截）
MSG_PROCUREMENT_FULL = (
    "我需要采购100台华硕笔记本电脑，用于办公室员工使用，"
    "预算50万元，需要在2026年3月底前完成交付，交付到北京总部，"
    "要求i7处理器16G内存，无特殊要求"
)  # 包含所有字段的完整采购消息
MSG_PROCUREMENT_PARTIAL = "我要采购一批服务器"     # 部分信息（触发反问）
MSG_EMBED_TEXT = "采购笔记本电脑办公室使用"          # 用于向量化测试

# --- 输出路径 ---
OUTPUT_DIR = "../data"

print("配置加载完成")
print(f"  AI Engine   : {AI_ENGINE_URL}")
print(f"  Session IDs : {SESSION_PROCUREMENT_1}, {SESSION_PROCUREMENT_2}")
print(f"  LLM 测试次数: {N_SAMPLES_LLM}")
````

````python
# ============================================================
# CELL 1: 依赖导入
# ============================================================

import httpx
import asyncio
import time
import json
import os
from datetime import datetime
from pathlib import Path
from typing import Dict, List, Optional, Tuple, Any

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
matplotlib.rcParams['axes.unicode_minus'] = False

Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
print("依赖加载完成")
````

````python
# ============================================================
# CELL 2: 测量工具函数
# ============================================================

def stats(data: List[float]) -> Dict:
    if not data:
        return {"mean": None, "p50": None, "p95": None, "max": None, "min": None, "n": 0}
    return {
        "mean": round(np.mean(data), 2),
        "p50":  round(np.percentile(data, 50), 2),
        "p95":  round(np.percentile(data, 95), 2),
        "max":  round(np.max(data), 2),
        "min":  round(np.min(data), 2),
        "n":    len(data),
    }


def measure_post_timing(url: str, payload: dict, n: int, warmup: int = 1,
                         label: str = "") -> Tuple[List[float], List[Any]]:
    """测量 POST 请求延迟，返回 (latencies_ms, responses)"""
    latencies = []
    responses = []
    for i in range(n + warmup):
        try:
            t0 = time.perf_counter()
            with httpx.Client(timeout=TIMEOUT_S) as client:
                r = client.post(url, json=payload)
            elapsed = (time.perf_counter() - t0) * 1000
            if i >= warmup:
                latencies.append(elapsed)
                responses.append(r.json() if r.status_code == 200 else None)
        except Exception as e:
            if i >= warmup:
                print(f"  {label} 失败: {e}")
    return latencies, responses


def measure_sse_phases(url: str, payload: dict, n: int = N_SAMPLES_LLM,
                        label: str = "") -> Dict:
    """
    测量 SSE 流式端点的各阶段时间:
    - t_connect: 服务器接受连接（200 响应头）
    - t_first_chunk: 第一个 content 块到达
    - t_total: 完整响应结束（complete 事件）
    """
    t_connects    = []
    t_first_chunks = []
    t_totals      = []
    chunk_counts  = []

    for i in range(n):
        try:
            t_start = time.perf_counter()
            t_connect = None
            t_first_chunk = None
            chunk_count = 0

            with httpx.Client(timeout=TIMEOUT_S) as client:
                with client.stream("POST", url, json=payload) as resp:
                    t_connect = (time.perf_counter() - t_start) * 1000

                    for line in resp.iter_lines():
                        if line.startswith("event:"):
                            event_type = line[6:].strip()
                        elif line.startswith("data:"):
                            try:
                                data = json.loads(line[5:].strip())
                                if "content" in data and t_first_chunk is None:
                                    t_first_chunk = (time.perf_counter() - t_start) * 1000
                                chunk_count += 1
                            except:
                                pass
                        elif line == "" and chunk_count > 0:
                            pass  # empty line separator

            t_total = (time.perf_counter() - t_start) * 1000

            t_connects.append(t_connect)
            if t_first_chunk:
                t_first_chunks.append(t_first_chunk)
            t_totals.append(t_total)
            chunk_counts.append(chunk_count)

            print(f"  [{i+1}/{n}] connect={t_connect:.0f}ms "
                  f"first_chunk={t_first_chunk:.0f}ms "
                  f"total={t_total:.0f}ms "
                  f"chunks={chunk_count}")

        except Exception as e:
            print(f"  [{i+1}/{n}] 失败: {e}")

    return {
        "t_connect":     stats(t_connects),
        "t_first_chunk": stats(t_first_chunks),
        "t_total":       stats(t_totals),
        "avg_chunks":    round(np.mean(chunk_counts), 1) if chunk_counts else 0,
        "raw": {
            "t_connects":    t_connects,
            "t_first_chunks": t_first_chunks,
            "t_totals":      t_totals,
        }
    }


print("工具函数定义完成")
````

````python
# ============================================================
# CELL 3: 基线测量 — AI Engine 无 Agent 开销
# ============================================================

print("=" * 60)
print("基线: AI Engine 纯路由开销（无 Agent 逻辑）")
print("=" * 60)

baseline_results = {}

# GET /health — 最轻量端点
print("\n  GET /health...")
health_data = []
for i in range(N_SAMPLES_FAST + WARMUP_ROUNDS):
    t0 = time.perf_counter()
    try:
        with httpx.Client(timeout=TIMEOUT_S) as c:
            r = c.get(f"{AI_ENGINE_URL}/health")
        elapsed = (time.perf_counter() - t0) * 1000
        if i >= WARMUP_ROUNDS:
            health_data.append(elapsed)
    except Exception as e:
        print(f"  错误: {e}")
baseline_results["GET /health"] = stats(health_data)
print(f"  mean={baseline_results['GET /health']['mean']}ms")

# POST /api/v1/embed — 纯 embedding，无 LLM
print("\n  POST /api/v1/embed (sentence-transformers, 无 LLM)...")
embed_data, _ = measure_post_timing(
    f"{AI_ENGINE_URL}/api/v1/embed",
    {"text": MSG_EMBED_TEXT},
    n=N_SAMPLES_FAST,
    warmup=WARMUP_ROUNDS,
    label="/embed"
)
baseline_results["POST /embed (cold=False)"] = stats(embed_data)
print(f"  mean={baseline_results['POST /embed (cold=False)']['mean']}ms")

# 测量同一文本重复 embed（验证 Redis 缓存效果）
print("\n  POST /api/v1/embed (同文本重复，测试缓存)...")
embed_repeat_data, _ = measure_post_timing(
    f"{AI_ENGINE_URL}/api/v1/embed",
    {"text": MSG_EMBED_TEXT},  # 与上面相同文本
    n=10,
    warmup=1,
    label="/embed (repeat)"
)
baseline_results["POST /embed (repeat)"] = stats(embed_repeat_data)
print(f"  mean={baseline_results['POST /embed (repeat)']['mean']}ms")

print("\n[基线摘要]")
for k, v in baseline_results.items():
    print(f"  {k:<40} mean={v['mean']}ms  p95={v['p95']}ms")
````

````python
# ============================================================
# CELL 4: [AGENT-1] GreeterAgent 意图识别（LLM call #1）
# 通过流式端点测量 TTFB = GreeterAgent LLM 调用时间
# ============================================================

print("=" * 60)
print("[AGENT-1] GreeterAgent 意图识别")
print("=" * 60)

greeter_results = {}

# 场景 A: 非采购意图 → GreeterAgent 拦截，不进入 RequirementAgent
# 此时 SSE first_chunk = GreeterAgent LLM TTFB + 转发消息开始时间
print(f"\n  场景 A: 非采购意图 '{MSG_NON_PROCUREMENT}'")
print(f"  (GreeterAgent 识别后直接返回，不调用 RequirementAgent)")

non_proc_results = []
for i in range(N_SAMPLES_LLM):
    sid = f"{SESSION_GREETER_ONLY}-{i}"
    t_start = time.perf_counter()
    t_first = None
    t_total = None
    try:
        with httpx.Client(timeout=TIMEOUT_S) as c:
            with c.stream(
                "POST",
                f"{AI_ENGINE_URL}/api/v1/requirement/stream",
                json={"message": MSG_NON_PROCUREMENT, "session_id": sid}
            ) as resp:
                for line in resp.iter_lines():
                    if line.startswith("data:"):
                        try:
                            d = json.loads(line[5:].strip())
                            if "content" in d and t_first is None:
                                t_first = (time.perf_counter() - t_start) * 1000
                        except:
                            pass
        t_total = (time.perf_counter() - t_start) * 1000
        non_proc_results.append({"t_first": t_first, "t_total": t_total})
        print(f"  [{i+1}/{N_SAMPLES_LLM}] first={t_first:.0f}ms total={t_total:.0f}ms")
    except Exception as e:
        print(f"  [{i+1}/{N_SAMPLES_LLM}] 错误: {e}")

greeter_t_first = [r["t_first"] for r in non_proc_results if r["t_first"]]
greeter_t_total = [r["t_total"] for r in non_proc_results]

greeter_results["non_procurement"] = {
    "t_first_chunk": stats(greeter_t_first),
    "t_total": stats(greeter_t_total),
    "description": "GreeterAgent LLM 调用 + 流式返回引导消息"
}

print(f"\n  GreeterAgent (非采购意图):")
print(f"    TTFB (first chunk): mean={stats(greeter_t_first)['mean']}ms")
print(f"    总计:               mean={stats(greeter_t_total)['mean']}ms")
````

````python
# ============================================================
# CELL 5: [AGENT-2] RequirementAgent 字段提取（LLM call #2）
# 测量包含完整信息时的字段提取时间（GreeterAgent + RequirementAgent）
# ============================================================

print("=" * 60)
print("[AGENT-2] RequirementAgent 字段提取（含完整采购信息）")
print("=" * 60)

requirement_results = {}

# 场景 B: 完整采购信息一次提供（GreeterAgent + RequirementAgent + 可能 ClerkAgent）
print(f"\n  场景 B: 完整采购消息（所有字段一次提供）")
print(f"  消息长度: {len(MSG_PROCUREMENT_FULL)} 字符")

full_stream_results = []
for i in range(N_SAMPLES_LLM):
    # 每次使用新 session 避免状态累积
    sid = f"{SESSION_PROCUREMENT_1}-{i}"
    t_start = time.perf_counter()
    t_first = None
    t_complete = None
    chunk_count = 0
    doc_received = False

    try:
        with httpx.Client(timeout=TIMEOUT_S) as c:
            with c.stream(
                "POST",
                f"{AI_ENGINE_URL}/api/v1/requirement/stream",
                json={"message": MSG_PROCUREMENT_FULL, "session_id": sid}
            ) as resp:
                current_event = None
                for line in resp.iter_lines():
                    if line.startswith("event:"):
                        current_event = line[6:].strip()
                    elif line.startswith("data:"):
                        try:
                            d = json.loads(line[5:].strip())
                            if "content" in d and t_first is None:
                                t_first = (time.perf_counter() - t_start) * 1000
                            if "doc" in d:
                                doc_received = True
                            chunk_count += 1
                        except:
                            pass
                    elif current_event == "complete":
                        t_complete = (time.perf_counter() - t_start) * 1000

        t_total = (time.perf_counter() - t_start) * 1000
        full_stream_results.append({
            "t_first": t_first,
            "t_total": t_total,
            "doc_received": doc_received,
            "chunks": chunk_count
        })
        print(f"  [{i+1}/{N_SAMPLES_LLM}] first={t_first:.0f}ms total={t_total:.0f}ms "
              f"doc={doc_received} chunks={chunk_count}")

    except Exception as e:
        print(f"  [{i+1}/{N_SAMPLES_LLM}] 错误: {e}")

full_t_first = [r["t_first"] for r in full_stream_results if r.get("t_first")]
full_t_total = [r["t_total"] for r in full_stream_results]

requirement_results["full_message_stream"] = {
    "t_first_chunk": stats(full_t_first),
    "t_total":       stats(full_t_total),
    "doc_received":  sum(1 for r in full_stream_results if r.get("doc_received")),
    "description":   "GreeterAgent + RequirementAgent (完整字段) + ClerkAgent"
}

print(f"\n  RequirementAgent (完整消息):")
print(f"    TTFB (first chunk): mean={stats(full_t_first)['mean']}ms")
print(f"    总计:               mean={stats(full_t_total)['mean']}ms")
print(f"    doc 生成率:         {requirement_results['full_message_stream']['doc_received']}/{N_SAMPLES_LLM}")
````

````python
# ============================================================
# CELL 6: 非流式 vs 流式对比
# 测量 POST /api/v1/requirement (非流式) 的 E2E 时间
# ============================================================

print("=" * 60)
print("非流式 vs 流式对比")
print("=" * 60)

nonstream_results = {}

# 非流式: 完整 Agent 链，返回完整响应
print("\n  POST /api/v1/requirement (非流式)...")
ns_latencies = []
for i in range(N_SAMPLES_LLM):
    sid = f"bench-ns-{_ts}-{i}"
    t0 = time.perf_counter()
    try:
        with httpx.Client(timeout=TIMEOUT_S) as c:
            r = c.post(
                f"{AI_ENGINE_URL}/api/v1/requirement",
                json={"message": MSG_PROCUREMENT_PARTIAL, "session_id": sid}
            )
        elapsed = (time.perf_counter() - t0) * 1000
        ns_latencies.append(elapsed)
        reply = r.json().get("reply", "")[:50] if r.status_code == 200 else "Error"
        print(f"  [{i+1}/{N_SAMPLES_LLM}] {elapsed:.0f}ms  reply='{reply}...'")
    except Exception as e:
        print(f"  [{i+1}/{N_SAMPLES_LLM}] 错误: {e}")

nonstream_results["partial_message"] = stats(ns_latencies)

# 流式 TTFB for partial message
print("\n  POST /api/v1/requirement/stream (流式) + 部分信息...")
stream_partial_results = []
for i in range(N_SAMPLES_LLM):
    sid = f"bench-sp-{_ts}-{i}"
    t_start = time.perf_counter()
    t_first = None
    try:
        with httpx.Client(timeout=TIMEOUT_S) as c:
            with c.stream(
                "POST",
                f"{AI_ENGINE_URL}/api/v1/requirement/stream",
                json={"message": MSG_PROCUREMENT_PARTIAL, "session_id": sid}
            ) as resp:
                for line in resp.iter_lines():
                    if line.startswith("data:"):
                        try:
                            d = json.loads(line[5:].strip())
                            if "content" in d and t_first is None:
                                t_first = (time.perf_counter() - t_start) * 1000
                                break  # 得到 TTFB 后可以退出
                        except:
                            pass
        if t_first:
            stream_partial_results.append(t_first)
            print(f"  [{i+1}/{N_SAMPLES_LLM}] TTFB={t_first:.0f}ms")
    except Exception as e:
        print(f"  [{i+1}/{N_SAMPLES_LLM}] 错误: {e}")

nonstream_results["stream_partial_ttfb"] = stats(stream_partial_results)

print("\n[对比总结]")
ns_mean = nonstream_results["partial_message"].get("mean")
st_mean = nonstream_results["stream_partial_ttfb"].get("mean")
print(f"  非流式 E2E:    mean={ns_mean}ms")
print(f"  流式 TTFB:     mean={st_mean}ms")
if ns_mean and st_mean:
    print(f"  TTFB vs E2E:   {st_mean/ns_mean*100:.1f}% (用户感知延迟改善)")
````

````python
# ============================================================
# CELL 7: Agent 编排流程静态分析
# 可视化 Agent 状态机 + 每个节点的时间分配
# ============================================================

print("=" * 60)
print("Agent 编排流程分析")
print("=" * 60)

# 基于测量数据估算各节点时间
# 注意: 以下时间是估算值，基于测量结果推导

greeter_llm_ms  = greeter_results.get("non_procurement", {}).get("t_first_chunk", {}).get("mean", 2000)
full_ttfb_ms    = stats(full_t_first).get("mean", 3000) if full_t_first else 3000
full_total_ms   = stats(full_t_total).get("mean", 8000) if full_t_total else 8000

# 推导:
# full_ttfb = greeter_llm + requirement_llm_ttfb
requirement_llm_estimate = max(full_ttfb_ms - greeter_llm_ms, 500)
# full_total = greeter + requirement + clerk
clerk_estimate = max(full_total_ms - full_ttfb_ms, 500)

# 已知固定开销
fastapi_routing_ms = baseline_results.get("GET /health", {}).get("mean", 5)
embed_ms           = baseline_results.get("POST /embed (cold=False)", {}).get("mean", 80)

agent_timeline = [
    # (节点名, 时间ms, 颜色, 说明)
    ("FastAPI 路由匹配",      fastapi_routing_ms,      "#95a5a6", "FastAPI 框架 + CORS"),
    ("Session 加载",          2.0,                     "#3498db", "内存 dict 查找，忽略不计"),
    ("[AGENT-1] GreeterAgent LLM", greeter_llm_ms,    "#e74c3c", "LLM 意图分类 call_llm(greeter_prompt)"),
    ("[AGENT-2] Req. 字段提取 LLM", requirement_llm_estimate, "#e67e22", "call_llm(requirement_prompt, return_json=True)"),
    ("[AGENT-2] category_agent",  5.0,                "#27ae60", "关键词匹配品类，无 LLM"),
    ("[AGENT-3] ClerkAgent LLM",  clerk_estimate,     "#9b59b6", "call_llm(clerk_prompt) 生成 PR 文档"),
    ("Session 写入",          2.0,                     "#3498db", "内存 dict 写入"),
    ("SSE 序列化 + 传输",      10.0,                   "#1abc9c", "EventSourceResponse 格式化"),
]

total_estimate = sum(t[1] for t in agent_timeline)
print(f"\n[Agent 链时间估算] (完整采购流程，含所有字段)")
print(f"  {'节点':<35} {'时间(ms)':>10} {'占比':>8} {'说明'}")
print("  " + "-" * 90)
for name, ms, color, desc in agent_timeline:
    pct = ms / total_estimate * 100
    print(f"  {name:<35} {ms:>10.0f} {pct:>7.1f}%  {desc}")
print(f"  {'合计':<35} {total_estimate:>10.0f} {'100%':>8}")

print(f"\n  实测 TTFB: {full_ttfb_ms:.0f}ms (包含 Greeter + Req LLM 调用)")
print(f"  实测 Total: {full_total_ms:.0f}ms (完整流程)")
print(f"  估算 Total: {total_estimate:.0f}ms (基于各节点推导)")
````

````python
# ============================================================
# CELL 8: LangGraph 图结构说明（静态分析）
# ============================================================

print("=" * 60)
print("LangGraph 图结构分析")
print("=" * 60)

print("""
LangGraph StateGraph (RequirementAgent.build_graph):

  Entry: extract_fields
    ↓
  Node: extract_fields
    - 调用 self.extractor.extract(user_text, domain)
    - 合并到 state["fields"]
    ↓
  Node: check_completeness
    - 调用 self.extractor.validate(fields, required_names)
    - 生成 next_question（配置驱动，不硬编码）
    ↓
  Conditional Edge (_route_after_check):
    - "complete" → END
    - "done"     → END

重要注意:
  - 流式路径 (run_requirement_agent_stream) 使用手动 if/elif 状态机，
    NOT 编译后的 LangGraph graph
  - LangGraph graph 在非流式路径 (RequirementAgent.process_message) 中使用
  - LangGraph 编译开销: 仅在 build_graph() 调用时发生（一次性，启动时）
  - 运行时 LangGraph 节点调度开销: ~1-5ms/节点（Python 函数调用）
""")

# 实际使用中的状态机流程（流式路径）
print("流式路径实际执行的状态机:")
print("""
  sessions[session_id]["current_step"]
  
  "greeter"  → GreeterAgent.route()
                 ↓ intent == "procurement"
              session["current_step"] = "requirement"
              
  "requirement" → RequirementAgent 字段提取
                    - 检查 missing 字段
                    - call_llm(requirement_prompt)     ← LLM call
                    - category_agent.identify_category() ← 关键词
                    - 更新 session["fields"]
                    ↓ is_complete == True
                  session["current_step"] = "clerk"
                  
  "clerk"     → 生成文档
                    - call_llm(clerk_prompt)           ← LLM call
                    - yield {"type": "doc", ...}
                  session["current_step"] = "done"
""")
````

````python
# ============================================================
# CELL 9: 综合可视化 — 3 个子图
# ============================================================

fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle("Agent 编排层追踪", fontsize=16, fontweight='bold', y=1.02)

COLORS = {
    "baseline": "#95a5a6",
    "greeter":  "#e74c3c",
    "require":  "#e67e22",
    "clerk":    "#9b59b6",
    "embed":    "#27ae60",
    "sse":      "#1abc9c",
}

# ── 子图 1: Agent 链时间瀑布图 ──────────────────────────────
ax1 = axes[0]
ax1.set_title("Agent 链时间分布\n（完整采购流程估算）", fontweight='bold')

node_names = [t[0] for t in agent_timeline]
node_vals  = [t[1] for t in agent_timeline]
node_cols  = [t[2] for t in agent_timeline]

bars = ax1.barh(range(len(agent_timeline)), node_vals, color=node_cols)
ax1.set_yticks(range(len(agent_timeline)))
ax1.set_yticklabels(node_names, fontsize=8)
ax1.set_xlabel("时间 (ms)")
for bar, val in zip(bars, node_vals):
    ax1.text(bar.get_width() + total_estimate * 0.01,
             bar.get_y() + bar.get_height()/2,
             f"{val:.0f}ms", va='center', fontsize=8)
ax1.set_xlim(0, max(node_vals) * 1.3)

# ── 子图 2: 流式 vs 非流式对比 ────────────────────────────
ax2 = axes[1]
ax2.set_title("流式 TTFB vs 非流式 E2E\n（用户感知延迟对比）", fontweight='bold')

stream_ttfb  = nonstream_results.get("stream_partial_ttfb", {}).get("mean", 0)
nonstream_e2e = nonstream_results.get("partial_message", {}).get("mean", 0)
full_e2e     = stats(full_t_total).get("mean", 0) if full_t_total else 0

labels = ["流式 TTFB\n(partial msg)", "非流式 E2E\n(partial msg)", "流式 Total\n(full msg)"]
values = [stream_ttfb, nonstream_e2e, full_e2e]
bar_colors = [COLORS["embed"], COLORS["require"], COLORS["clerk"]]

bars = ax2.bar(range(len(labels)), values, color=bar_colors)
ax2.set_xticks(range(len(labels)))
ax2.set_xticklabels(labels, fontsize=9)
ax2.set_ylabel("时间 (ms)")
for bar, val in zip(bars, values):
    ax2.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(values)*0.01,
             f"{val:.0f}ms", ha='center', va='bottom', fontsize=10, fontweight='bold')

ax2.set_title(f"流式 TTFB vs 非流式 E2E\n用户等待感知减少: {(1-stream_ttfb/nonstream_e2e)*100:.0f}%" 
              if nonstream_e2e > 0 else "流式 TTFB vs 非流式 E2E",
              fontweight='bold')

# ── 子图 3: Agent 占比饼图 ──────────────────────────────────
ax3 = axes[2]
ax3.set_title("完整采购流程\n各 Agent 时间占比", fontweight='bold')

pie_data = [
    ("FastAPI 框架",     fastapi_routing_ms,      COLORS["baseline"]),
    ("GreeterAgent LLM", greeter_llm_ms,          COLORS["greeter"]),
    ("字段提取 LLM",      requirement_llm_estimate, COLORS["require"]),
    ("品类识别",          5.0,                      COLORS["embed"]),
    ("ClerkAgent LLM",   clerk_estimate,           COLORS["clerk"]),
    ("其他",              12.0,                     COLORS["sse"]),
]

pie_vals  = [d[1] for d in pie_data]
pie_labels = [f"{d[0]}\n{d[1]:.0f}ms" for d in pie_data]
pie_cols  = [d[2] for d in pie_data]

wedges, texts, autotexts = ax3.pie(
    pie_vals, labels=pie_labels, colors=pie_cols,
    autopct='%1.1f%%', startangle=90, pctdistance=0.8,
    textprops={'fontsize': 8}
)
for autotext in autotexts:
    autotext.set_fontsize(8)

plt.tight_layout()
ts = datetime.now().strftime("%Y%m%d_%H%M")
save_path = f"{OUTPUT_DIR}/agent-orchestration-trace_{ts}.png"
plt.savefig(save_path, dpi=150, bbox_inches='tight')
plt.show()
print(f"图表已保存: {save_path}")
````
