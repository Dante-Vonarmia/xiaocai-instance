# 01 - E2E 延迟测试

**目的**: 测量从用户发送消息到系统返回响应的完整链路延迟  
**测试路径**: 测试客户端 → API Gateway (8001) → AI Engine (8002) → LLM Provider  
**输出**: P50/P90/P95/Max 延迟统计 + 可视化图表 + CSV 结果文件

**前提**: 系统已启动（`./start.sh` 或 `./xiaocai.sh`）

## 0. 配置参数（修改此处进行实验）

```python
# =============================================
# 实验配置 - 按需修改
# =============================================

# 服务端点
AI_ENGINE_URL = "http://localhost:8002"
API_GATEWAY_URL = "http://localhost:8001"

# 测试轮数（建议 10-20 轮获得稳定统计）
N_ITERATIONS = 10

# 请求超时（秒）
REQUEST_TIMEOUT = 120

# 测试消息集（可扩展）
TEST_MESSAGES = [
    {
        "id": "simple_greeting",
        "label": "简单问候",
        "text": "你好"
    },
    {
        "id": "partial_requirement",
        "label": "部分信息需求",
        "text": "我需要采购一批服务器"
    },
    {
        "id": "full_requirement",
        "label": "完整需求描述",
        "text": "我要为研发部门采购100台Dell PowerEdge R740服务器，配置双路Intel Xeon Gold 6248R处理器，512GB内存，用于AI模型训练，预算800万元，需要在2026年3月31日前交货到北京朝阳区，要求原厂5年保修"
    }
]

# 结果保存路径
import os
from datetime import datetime
RESULTS_DIR = "../data"
os.makedirs(RESULTS_DIR, exist_ok=True)
RESULTS_FILE = f"{RESULTS_DIR}/e2e_latency_{datetime.now().strftime('%Y%m%d_%H%M')}.csv"

print(f"配置加载完成")
print(f"AI Engine: {AI_ENGINE_URL}")
print(f"API Gateway: {API_GATEWAY_URL}")
print(f"测试轮数: {N_ITERATIONS}")
print(f"测试消息数: {len(TEST_MESSAGES)}")
print(f"结果文件: {RESULTS_FILE}")
```

## 1. 依赖导入

```python
import httpx
import time
import json
import uuid
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from datetime import datetime

# 中文显示
matplotlib.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'DejaVu Sans']
matplotlib.rcParams['axes.unicode_minus'] = False

print("依赖导入完成")
```

## 2. 服务连通性检查

```python
def check_service(name: str, url: str, timeout: float = 5.0) -> dict:
    """检查服务健康状态"""
    try:
        resp = httpx.get(f"{url}/health", timeout=timeout)
        return {
            "service": name,
            "url": url,
            "status": "ok" if resp.status_code == 200 else "error",
            "latency_ms": resp.elapsed.total_seconds() * 1000,
            "detail": resp.json() if resp.status_code == 200 else resp.text
        }
    except Exception as e:
        return {
            "service": name,
            "url": url,
            "status": "unreachable",
            "latency_ms": None,
            "detail": str(e)
        }

# 检查所有服务
services = [
    ("AI Engine", AI_ENGINE_URL),
    ("API Gateway", API_GATEWAY_URL),
]

print("=" * 50)
print("服务连通性检查")
print("=" * 50)

all_ok = True
for name, url in services:
    result = check_service(name, url)
    status_icon = "OK" if result["status"] == "ok" else "FAIL"
    latency = f"{result['latency_ms']:.1f}ms" if result['latency_ms'] else "N/A"
    print(f"[{status_icon}] {name:15s} {url:35s} {latency}")
    if result["status"] != "ok":
        print(f"       错误: {result['detail']}")
        all_ok = False

print("=" * 50)
if all_ok:
    print("所有服务正常，可以开始测试")
else:
    print("部分服务不可用，请先启动服务")
```

## 3. 单次 E2E 请求测试（快速验证）

```python
def send_requirement_request(message: str, session_id: str = None) -> dict:
    """
    发送需求提取请求到 AI Engine
    直接调用 AI Engine /api/v1/requirement 端点
    """
    if not session_id:
        session_id = str(uuid.uuid4())

    payload = {
        "message": message,
        "session_id": session_id,
        "stream": False
    }

    start = time.perf_counter()
    try:
        resp = httpx.post(
            f"{AI_ENGINE_URL}/api/v1/requirement",
            json=payload,
            timeout=REQUEST_TIMEOUT
        )
        elapsed = (time.perf_counter() - start) * 1000  # ms

        return {
            "success": resp.status_code == 200,
            "status_code": resp.status_code,
            "latency_ms": elapsed,
            "response": resp.json() if resp.status_code == 200 else None,
            "error": None if resp.status_code == 200 else resp.text
        }
    except Exception as e:
        elapsed = (time.perf_counter() - start) * 1000
        return {
            "success": False,
            "status_code": None,
            "latency_ms": elapsed,
            "response": None,
            "error": str(e)
        }


# 快速验证 - 发送一条测试消息
print("发送测试请求...")
test_msg = TEST_MESSAGES[1]["text"]
result = send_requirement_request(test_msg)

print(f"状态: {'成功' if result['success'] else '失败'}")
print(f"延迟: {result['latency_ms']:.0f} ms")
if result["response"]:
    print(f"响应预览: {str(result['response'])[:200]}...")
if result["error"]:
    print(f"错误: {result['error']}")
```

## 4. 批量延迟测试

```python
import time as time_module

all_results = []

print(f"开始批量测试: {N_ITERATIONS} 轮 × {len(TEST_MESSAGES)} 种消息")
print("=" * 60)

for i in range(N_ITERATIONS):
    for msg_cfg in TEST_MESSAGES:
        session_id = str(uuid.uuid4())
        result = send_requirement_request(msg_cfg["text"], session_id)

        record = {
            "iteration": i + 1,
            "message_id": msg_cfg["id"],
            "message_label": msg_cfg["label"],
            "latency_ms": result["latency_ms"],
            "success": result["success"],
            "timestamp": datetime.now().isoformat()
        }
        all_results.append(record)

        status = "OK" if result["success"] else "FAIL"
        print(f"[{status}] 轮{i+1:02d} | {msg_cfg['label']:12s} | {result['latency_ms']:7.0f}ms")

    # 轮次间隔，避免限流
    if i < N_ITERATIONS - 1:
        time_module.sleep(1)

df = pd.DataFrame(all_results)
print("\n测试完成")
print(f"总请求数: {len(df)}")
print(f"成功率: {df['success'].mean()*100:.1f}%")
```

## 5. 延迟统计分析

```python
# 只分析成功请求
df_ok = df[df["success"] == True].copy()

print("=" * 60)
print("E2E 延迟统计 (ms)")
print("=" * 60)

# 按消息类型分组统计
stats = df_ok.groupby("message_label")["latency_ms"].agg([
    ("P50", lambda x: np.percentile(x, 50)),
    ("P90", lambda x: np.percentile(x, 90)),
    ("P95", lambda x: np.percentile(x, 95)),
    ("Max", "max"),
    ("Min", "min"),
    ("Mean", "mean"),
    ("Count", "count")
]).round(0)

print(stats.to_string())

print("\n" + "=" * 60)
print("整体统计")
print("=" * 60)
overall = df_ok["latency_ms"]
print(f"P50:  {np.percentile(overall, 50):7.0f} ms")
print(f"P90:  {np.percentile(overall, 90):7.0f} ms")
print(f"P95:  {np.percentile(overall, 95):7.0f} ms")
print(f"Max:  {overall.max():7.0f} ms")
print(f"Min:  {overall.min():7.0f} ms")
print(f"Mean: {overall.mean():7.0f} ms")
```

## 6. 可视化

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"E2E 延迟分布 (n={len(df_ok)})", fontsize=14)

# 图1: 按消息类型的箱线图
ax1 = axes[0]
df_ok.boxplot(column="latency_ms", by="message_label", ax=ax1)
ax1.set_title("按消息类型延迟分布")
ax1.set_xlabel("消息类型")
ax1.set_ylabel("延迟 (ms)")
ax1.tick_params(axis='x', rotation=15)

# 图2: 整体延迟直方图
ax2 = axes[1]
ax2.hist(df_ok["latency_ms"], bins=20, color="steelblue", edgecolor="white", alpha=0.8)
ax2.axvline(np.percentile(df_ok["latency_ms"], 50), color="green", linestyle="--", label=f"P50={np.percentile(df_ok['latency_ms'], 50):.0f}ms")
ax2.axvline(np.percentile(df_ok["latency_ms"], 90), color="orange", linestyle="--", label=f"P90={np.percentile(df_ok['latency_ms'], 90):.0f}ms")
ax2.axvline(np.percentile(df_ok["latency_ms"], 95), color="red", linestyle="--", label=f"P95={np.percentile(df_ok['latency_ms'], 95):.0f}ms")
ax2.set_title("整体延迟直方图")
ax2.set_xlabel("延迟 (ms)")
ax2.set_ylabel("频次")
ax2.legend()

plt.tight_layout()
plt.savefig(f"{RESULTS_DIR}/e2e_latency_chart_{datetime.now().strftime('%Y%m%d_%H%M')}.png", dpi=150, bbox_inches="tight")
plt.show()
print("图表已保存")
```

## 7. 保存结果

```python
df.to_csv(RESULTS_FILE, index=False, encoding="utf-8")
print(f"结果已保存: {RESULTS_FILE}")
print(f"共 {len(df)} 条记录")
df.head()
```
